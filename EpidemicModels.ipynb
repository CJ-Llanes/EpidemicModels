# Part 1

Mathematical modeling of epidemics is a technique that uses mathematical language and tools to explain and predict their evolution. Here, as in many other fields, generating a mathematical model can serve to understand the essential aspects of the interaction with the infectious agent, determine what data is indicative, have a theoretical framework where virtual experiments can be performed, etc.

In an epidemic, the disease can infect individuals randomly. However, the law of large numbers indicates that the number of infected individuals becomes increasingly predictable, as population size increases. For this reason, for large populations, it is appropriate to model evolution through differential equations. On the contrary, for small populations the fluctuations can be considerable, the probability is present and stochastic modeling should be used, with random variables.

Suppose a population of size S initially healthy, in which a certain number I of infected is introduced. The object of study of most of these models is constituted by individuals, in a figurative and mathematically treatable sense, linked to their environment generally constituted by the closest neighbors. The interaction with their neighbors is modeled mathematically in an abstract way and stripped of all psychological or social aspects, and is associated with the fact that it is the spatial proximity between two individuals that makes the transmission of the disease between them more likely, in case of be contagious Thus, the spatial distribution will determine which is the group of individuals to which a subject may transmit or from whom the disease may be spread. There are models that contemplate immobile individuals, other subjects that migrate displacing the infection. During the infectious process and depending on the type of disease, individuals may go through all or some of the following stages:

* Susceptible (S), a state in which the individual can be infected by another agent who is infected
* Infected (I), a state during which the individual is infected and can also infect
* Removed (R), or recovered, state during which the individual can neither be infected by having acquired immunity (temporary or permanent) or affect (by having recovered or having passed the contagious stage of the disease).


# SIR Model

The susceptible-infected-removed model, related to diseases that confer permanent immunity and a typical cycle includes all three stages. This does not mean that all individuals in a population must go through them, some will not be infected and remain healthy, that is, always in an S state, others will be artificially immunized by vaccination, or some other method and will become R without having been infected. It is precisely the interest of the model to take into account all these possibilities and try to predict the behavior of an epidemic.

# SIS Model

The susceptible-infected-susceptible model; they are used in cases where the disease does not confer immunity and the individual goes from being infected to susceptible again, skipping stage R.
A model may or may not take into account the vital dynamics of the population (births, deaths, migratory movements) depending mainly on the times involved in the described process, determined mainly by the characteristics of the disease and the society under study.

# SI Model

The SI model was introduced in 1927 by Kermack [1].

In this model, during the course of an epidemics, a node is allowed to change its status only from Susceptible (S) to Infected (I).

The model is instantiated on a graph having a non-empty set of infected nodes.

SI assumes that if, during a generic iteration, a susceptible node comes into contact with an infected one, it becomes infected with probability β: once a node becomes infected, it stays infected (the only transition allowed is S→I).

```python
import scipy.integrate
import numpy
import matplotlib.pyplot as plt
from past.builtins import xrange
```

```python
def SIR(y, t, beta, gamma):
    S, I, R = y
    
    dS_dt = -beta*S*I 
    dI_dt = beta*S*I - gamma*I
    dR_dt = gamma*I
    
    return ([dS_dt, dI_dt, dR_dt])

S0 = 0.9
I0 = 0.2
R0 = 0.0
beta = 0.35
gamma = 0.15

#time vector
t = numpy.linspace(0,100,190000)

#result
solu = scipy.integrate.odeint(SIR, [S0,I0,R0], t, args =(beta, gamma))
solu = numpy.array(solu)
```

```python
def SIS(y, t, beta, gamma):
    s, i = y
    
    dS_dt = -beta*s*i + gamma * i
    dI_dt = beta*s*i - gamma*i
    
    

    return ([dS_dt, dI_dt,])
#N = 10000
s0 = 1
i0 = 1e-6
beta = .4247
gamma = 0.14286





#time vector
t2 = numpy.linspace(0,100,1000)


#result
solution = scipy.integrate.odeint(SIS, [s0,i0], t2, args =(beta, gamma))
solution = numpy.array(solution)
```

# Part 4

Networks and the epidemiology of directly transmitted infectious diseases are fundamentally linked. The foundations of epidemiology and early epidemiological models were based on population wide random-mixing, but in practice each individual has a finite set of contacts to whom they can pass infection; the ensemble of all such contacts forms a ‘mixing network’. Knowledge of the structure of the network allows models to compute the epidemic dynamics at the population scale from the individual-level behaviour of infections. Therefore, characteristics of mixing networks—and how these deviate from the random-mixing norm—have become important applied concerns that may enhance the understanding and prediction of epidemic patterns and intervention measures.

Epidemiological models are mathematical models. We understand a mathematical model as a collection of relationships between variables. The latter are ways of quantifying relevant aspects of reality, and the relationships between them are intended to capture either the dynamics or the structure of the phenomenon of interest. Usually, this dynamic / structure presents non-trivial challenges to our understanding, either because it is not directly or easily observable, or because it is very complex and difficult to collect and manipulate the multitude of factors involved. Mathematical models come to the rescue to allow us to describe, in a simplified way, the dynamics / structure of the phenomenon that interests us. A model is a simplification that allows us to have an intelligible insight into what we want to understand. This leads to modeling being a process of balance between two forces. On the one hand, the relevance, which seeks to include in the model as many relevant factors as possible; and, on the other, simplicity, which seeks to concentrate on the main aspects that allow to gain understanding about the central tendency of the dynamics, or the basic structure skeleton, and not to overwhelm with details that obscure them.

```python
# NetworkX for network representation
import networkx as nx
# Some handy mathematical functions
import math
# Random numbers from numpy
import numpy as np
import numpy.random as rnd
# Important elements of matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as clr
import matplotlib.cm as cmx
# Plot all graphics in-line in the notebook
%matplotlib inline
```

```python
SPREADING_SUSCEPTIBLE = 'S'
SPREADING_INFECTED = 'I'
SPREADING_RECOVERED = 'R'
def spreading_init( g ):
    for i in g.node.keys():
        er.node[i]['state'] = SPREADING_SUSCEPTIBLE

        
def spreading_seed( g, pSick ):
    for i in g.node.keys():
        if(rnd.random() <= pSick):
            er.node[i]['state'] = SPREADING_INFECTED


            
def spreading_make_sir_model( pInfect, pRecover ):
    def model( g, i ):
        if g.node[i]['state'] == SPREADING_INFECTED:
            for m in g.neighbors(i):
                if g.node[m]['state'] == SPREADING_SUSCEPTIBLE:
                    if rnd.random() <= pInfect:
                        g.node[m]['state'] = SPREADING_INFECTED
                        
            # recover with probability pRecover
            if rnd.random() <= pRecover:
                g.node[i]['state'] = SPREADING_RECOVERED
    return model


def spreading_step( g, model ):
    for i in g.node.keys():
        model(g, i)
def spreading_run( g, model, iter ):
    for i in xrange(iter):
        spreading_step(g, model)
n = 1000
er = nx.erdos_renyi_graph(n,0.04)
fig = plt.figure(figsize=(10, 8))
# some visual styling for the figure
ax = fig.gca()
ax.grid(False) # no grid
ax.get_xaxis().set_ticks([]) # no ticks on the axes
ax.get_yaxis().set_ticks([])
# run the spring layout algorithm over the network
pos = nx.spring_layout(er, iterations = 50, k = 2/math.sqrt(n))
# draw the network using the computed positions for the nodes
nx.draw_networkx_edges(er, pos, width = 1, alpha = 0.4)
nx.draw_networkx_nodes(er, pos, node_size=100, alpha = 1, linewidths = 0.5)
plt.show()
```

```python
# initialise with 5% sick people
spreading_init(er)
spreading_seed(er, 0.05)
model = spreading_make_sir_model(0.3, 0.05)
# run SIR model dynamics over the network
spreading_run(er, model, 100)
infected = [ v for (v, attr) in er.nodes(data = True) if attr['state'] == SPREADING_RECOVERED ]
print (float(len(infected)) / n)
```
